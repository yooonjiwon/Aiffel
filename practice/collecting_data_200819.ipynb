## Internet이란?
- TCP/IP(Transmission Control Protocol / Internet Protocol)이라는 통신 프로토콜을 이용해 정보를 주고 받는 컴퓨터 네트워크
- 1960-70년도에 두 연구소간 네트워크를 연결한 것이 인터넷의 시초
- TCP-IP를 기반으로 HTTP 프로토콜 이용할 수 있는 소프트웨어: 웹 브라우저(크롬, 파이어폭스, 사파리)
- 1995년 10월 24일에 처음으로 인터넷이라는 용어를 사용

## Socket 통신
- 소켓: socket을 통해 인터넷 네트워크에 접속하고 데이터를 가져오게 됨

```python
# socket 객체를 생성한 뒤 구글에 접속해서 데이터 받아옴.
import socket
with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:     # 소켓 s를 하나 열어서
    s.connect(('www.google.com', 80))          # www.google.com 이라는 호스트의 80번 포트에 접속합니다. 
    cmd = 'GET http://www.google.com HTTP/1.0\r\n\r\n'.encode() 
    s.send(cmd)                 # 접속이 되면 GET xxxxx 형식의 HTTP 프로토콜로 된 명령을 전송합니다. 
    data = s.recv(1024)    # 그리고 1024바이트만큼의 HTTP 응답을 수신합니다. 
print('Received', data)  # 어떤 응답을 수신했는지 볼까요?
```

## 인터넷 프로토콜: TCP/IP
- TCP: Transmission Control Protocol -> 소켓 포트 단위의 송수신 프로토콜   
=> Transport 레이어에서 데이터의 흐름과 정확성 여부를 관리   
=> 여러 컴퓨터는 각각 다른 포트 번호의 소켓 가지고 있음.   
=> 포트번호 기준으로 target으로 정확히 전송될 수 있도록 하는 역할을 함.   


- IP: Internet Protocol -> 컴퓨터마다 주어지는 IP 주소 단위의 송수신 프로토콜
=> 패킷을 목적지까지 전달.   
=> 목적지는 IP 주소를 할당 받음.   
=> 주소로 데이터가 정확히 전송되도록 책임짐

## HTTP
- request, respone하는 서버/클라이언트 모델을 따름


### Request Method
- GET: read, 정보를 요청
- HEAD: 헤더 정보만 요청
- POST: create, 정보를 생성, 변형 할때 사용
- PUT: update, 정보를 수정
- DELETE: delete, 정보를 삭제
- OPTIONS : 사용가능한 메소드 확인

### Response Code
- 1xx: 조건부 응답
- 2xx: 성공
- 3xx: 리다이렉션 완료
- 4xx: 요청 오류(클라이언트에 문제가 있다는 뜻!)
- 5xx: 서버 오류(서버에 문제가 있다는 뜻!)   
ex) 200은 페이지 요청 정상적 완료, 404는 페이지 접속할 수 없을 때

### Google Crawler

```python
# Web Crawler
import urllib
def download(url):
    try:
        html = urllib.request.urlopen(url).read()
    except (URLError, HTTPError, ContentTooShortError) as e:
        print('Download error', e.reason)
        html = None
    return html
```

```python
download('https://www.google.com')
```

### 네이버 환율 정보 크롤링

```python
# 200은 응답에 성공했다는뜻
import requests
url = 'https://m.stock.naver.com/marketindex/index.nhn'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.html.body
```

```python
# 국가 추출
country = []
country_1 = soup.findAll('strong')
country_2 = soup.findAll('span', {'class':'stock_item'})
for c in country_1:
    country.append(c.text)
for c in country_2:
    country.append(c.text)
country.remove('컨텐츠 제공업체')
country.remove('증권')
country.remove('달러인덱스')
```

```python
#원하는 데이터 추출 - 환율정보
price = []
price_0 = soup.findAll('span', {"class":"stock_price"})
for p in price_0:
    price.append(p.text)
price
```

```python
#데이터 정렬, (pandas)
import pandas as pd
data = {}
for i in range(len(country)):
    data[country[i]] = price[i]

pd.Series(data)
```

- 쿠키: HTTP 헤더를 기반으로 이루어지는 데이터. (key-value)
- 세션: 쿠키로 데이터 저장.
- 쿠키와 세션의 차이점: 
- 1) 쿠키는 로컬에, 세션은 로컬+서버에 저장
- 2) 쿠키는 탈취 변조 가능, 세션은 상대적으로 안전
- 3) Lifecycle: 쿠키는 브라우저를 종료해도 파일로 남아있고, 세션은 삭제됨.
- 4) 속도: 쿠키는 빠른편이고, 세션은 느린편
